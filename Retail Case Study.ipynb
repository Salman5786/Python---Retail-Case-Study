## 1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure to keep all customers who have done transactions with us and select the join type accordingly.

```python
Customer_Trans = pd.merge(left = Customer, right = Transaction, left_on = 'customer_Id', right_on = 'cust_id', how = 'right')
Customer_Trans
```

```python
Customer_Final = pd.merge(left = Customer_Trans, right = Product_Heirarchy, left_on = ['prod_subcat_code', 'prod_cat_code'], right_on = ['prod_sub_cat_code', 'prod_cat_code'], how = 'left')
Customer_Final
```

## 6. Which City code has the maximum customers and what was the percentage of customers from that city?

```python
Max_Cust = Customer_Final.pivot_table(index = 'city_code', values = 'customer_Id', aggfunc = 'count').rename(columns = {'customer_Id' : 'Count of Cust'}).sort_values(by = 'Count of Cust', ascending = False)
Max_Cust['Percentage'] = Max_Cust.div(Max_Cust.sum(axis = 0)/100)
Max_Cust.head(1)
```

## 11. For all customers aged between 25 - 35, find out:

### a. What was the total amount spent for “Electronics” and “Books” product categories?

```python
Customer_Final.DOB = pd.to_datetime(Customer_Final.DOB, format = '%d-%m-%Y')
Max_tran_date = Customer_Final.loc[:, ['customer_Id', 'DOB', 'tran_date']].groupby('customer_Id').max()
Max_tran_date['Age'] = pd.DatetimeIndex(Max_tran_date['tran_date']).year - pd.DatetimeIndex(Max_tran_date['DOB']).year
```

```python
Customer_Final = pd.merge(left = Customer_Final, right = Max_tran_date, on = ['customer_Id', 'DOB'], how = 'left', suffixes = ('_', '_max'))
```

```python
total_amt_spent_a = Customer_Final.loc[(Customer_Final.Age >= 25) & (Customer_Final.Age <= 35) & (Customer_Final.prod_cat.isin(['Electronics', 'Books'])), 'total_amt'].sum()
```

```python
print('For all customers aged between 25-35, the total amount spent for Electronics and Books product categories is', total_amt_spent_a)
```

### b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?

```python
total_amt_spent_b = Customer_Final.loc[(Customer_Final.Age >= 25) & (Customer_Final.Age <= 35) & (Customer_Final.tran_date_ >= '2014-01-01') & (Customer_Final.tran_date_ <= '2014-03-01'), 'total_amt'].sum()
```

```python
print('For all customers aged between 25-35, the total amount spent by these customers between 1stJan,2014 to 1stMar,2014 is', total_amt_spent_b)
```
